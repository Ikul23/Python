```python
import pandas as pd
import random

# Создаем DataFrame
lst = ['robot'] * 10 + ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI': lst})

# Применяем one-hot кодирование без использования get_dummies
one_hot_encoded = pd.get_dummies(data['whoAmI'], prefix='whoAmI')

# Добавляем закодированные столбцы к исходному DataFrame
data_encoded = pd.concat([data, one_hot_encoded], axis=1)

# Удаляем исходный столбец 'whoAmI', так как он больше не нужен
data_encoded = data_encoded.drop('whoAmI', axis=1)

# Выводим результат
data_encoded.head()
```
